## Datasets
### Get Data

```python
main_path = os.path.dirname(os.getcwd())
file = '\\DADOS\\WDO$N_15M.csv'
file_path = main_path + file
df_wdo_15 = pd.read_csv(file_path,infer_datetime_format=True)
```

```python
main_path = os.path.dirname(os.getcwd())
file = '\\DADOS\\WIN$N_15M.csv'
file_path = main_path + file
df_win_15 = pd.read_csv(file_path,infer_datetime_format=True)
```

```python
dataset_names = ["WDO_15", "WIN_15"]
datasets = [df_wdo_15,df_win_15]
```

```python
train_split = 2000
test_length = 200
```

### Visualization

```python
fig, ax = plt.subplots(nrows=3, ncols=2, figsize=[10,5])

for count,dataset_name in enumerate(dataset_names):
    dataset = list(datasets[count].close)
    dataset_diff = tdiff.apply(dataset)
    dataset_boxcox = boxcox.apply(dataset)
    
    ax[0][count].plot(dataset)
    ax[1][count].plot(dataset_diff)
    ax[2][count].plot(dataset_boxcox)
    ax[0][count].set_title(dataset_name)
```

### Statistics

```python
# from statsmodels.tsa.stattools import adfuller
rows =[]

for count,dataset_name in enumerate(dataset_names):
    row = [dataset_name]
    dataset = list(datasets[count].close)
    result = adfuller(dataset)
    row.extend([result[0],result[1]])
    row.extend([value for key, value in result[4].items()])
    rows.append(row)
    
pd.DataFrame(rows,columns=['Dataset','ADF Statistic','p-value','Cr. Val. 1%','Cr. Val. 5%','Cr. Val. 10%'])
```
